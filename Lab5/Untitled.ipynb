```python
from datetime import datetime
from packaging import version

import tensorflow as tf
from tensorflow import keras
tf.debugging.experimental.enable_dump_debug_info('./logs/', tensor_debug_mode="FULL_HEALTH", circular_buffer_size = -1)
from keras import backend as K
import numpy as np
print("Tensorflow version: ", tf.__version__)
assert version.parse(tf.__version__).release[0] >= 2, \
    "Invalid tensorflow version, need tensorflow 2.0 or higher"
```

```python
data_size = 10000
train_percentage = 0.8
train_size = int(data_size * train_percentage)

x = np.linspace(-1, 2, data_size)
np.random.shuffle(x)
y = 2.5 * x + 10 + np.random.normal(0, 0.05, (data_size, ))

x_train, y_train = x[:train_size], y[:train_size]
x_test, y_test = x[train_size:], y[train_size:]
```

```python
logdir = "logs/scalars/" + datetime.now().strftime("%Y%m%d-%H%M%S")
tensorboard_callback = keras.callbacks.TensorBoard(
    log_dir=logdir,
    write_images=False,
    embeddings_freq=0
)
model = keras.models.Sequential([
    keras.layers.Dense(16, input_dim=1),
    keras.layers.Dense(1),
])

model.compile(
    loss='mse',
    optimizer=keras.optimizers.SGD(learning_rate=0.2),
)

print("Training ... With default parameters, this takes less than 10 seconds.")
training_history = model.fit(
    x_train,
    y_train,
    batch_size=train_size,
    verbose=1, 
    epochs=20,
    validation_data=(x_test, y_test),
    callbacks=[tensorboard_callback],
)

print("Average test loss: ", np.average(training_history.history['loss']))
```
